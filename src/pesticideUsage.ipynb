```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Load the dataset (assuming you have a CSV file with 'datetime' and 'temperature' columns)
data = pd.read_csv("../datasets/pesticide_usage_in_uk.csv")

data['suryear'] = data['suryear']
x = data[['suryear']]
data['kg'] = data['kg'].astype(float)
y = data['kg']

# # Initialize and fit the linear regression model
model = LinearRegression()
model.fit(x, y)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Pesticides Used(*10 million kg)')
plt.scatter(data.suryear,data.kg,  marker='.', label='Pesticides Used')
plt.plot(data[['suryear']], model.predict(x.values), color='red', linewidth=1, label='Linear regression')
plt.legend(loc=2)

# Save and display the plot
plt.savefig('../output-results/pesticideUsage.png')
plt.show()
```
